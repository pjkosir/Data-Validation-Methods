```python
# importing all necessary libraries and functions
import pandas as pd
import os
import matplotlib.pyplot as plt
import statistics
```

```python
# This function verifies that all the data is within a the specified range and
# counts the number of times that each variable is out of this range
def domain_check(self,Data_set,variable):
    problem = False
    toolow = 0
    toohigh = 0
    columns, rows = Data_set.shape
    for i in range(columns):
        if (Data_set[i][variable] < self.Domain_lower[variable]):
            toolow = toolow + 1
        elif (Data_set[i][variable] > self.Domain_upper[variable]):
            toohigh = toohigh + 1
    if (toolow > 0):
        print("There are",toolow,"values too low in",self.variable_names[variable])
        problem = True
    if (toohigh > 0):
        print("There are",toohigh,"values too high in",self.variable_names[variable])
        problem = True
    return problem
```

```python
# This funciton counts the number of times each variable continues increasing or decreasing
# and records this as a percentage called the momentum of that variable. It then verifies 
# that this value is within a given range which is unique to each variable. 
def momentum_check(self,Data_set,variable):
    problem = False
    momentum = 0 
    roc = get_roc(Data_set,variable)
    columns, rows = Data_set.shape
    for i in range(columns - 2):
        if ((roc[i] > 0) and (roc[i+1] > 0)):
            momentum = momentum + 1
        elif ((roc[i] < 0) and (roc[i+1] < 0)):
            momentum = momentum + 1
    mv = momentum / (columns - 2)
    if (mv > self.momentum_upper[variable]):
        print("The mometum value for",self.variable_names[variable],"is too high with a value of", mv)
    elif (mv < self.momentum_lower[variable]):
        print("The mometum value for",self.variable_names[variable],"is too low with a value of", mv)
        problem = True
    return problem
```

```python
# This function counts the number of times each of the three variables moves in the 
# same direction (increase or decrease). It then computes this as percentage called
# the agreement level and checks to make sure this value is within a given range.
def agreement_check(self,Data_set):
    problem = False
    agreement = 0
    columns, rows = Data_set.shape
    roc = []
    for i in range(rows):
        temp = get_roc(Data_set,i)
        roc.append(temp)
    for i in range(columns - 1):
        pos = 0
        neg = 0
        for j in range(rows):
            if (roc[j][i] > 0):
                pos = pos + 1
            elif (roc[j][i] < 0):
                neg = neg + 1
        if (pos >= self.agree_num):
            agreement = agreement + 1
        elif (neg >= self.agree_num):
            agreement = agreement + 1
    agreement_level = agreement / columns
    if (agreement_level < self.agree_lower) or (agreement_level > self.agree_upper):
        print("Alert: Agreement level out of range with a value of",agreement_level)
        problem = True
    return problem
```

```python
# This function gets the distribution for a given data set
# This is an intermediate function
def get_distribution(Data_set):
    distribution = [0] * 13
    regions = [-3,-2.5,-2,-1.5,-1,-.5,.5,1,1.5,2,2.5,3]
    cutoffs = [0] * 12
    mean = statistics.mean(Data_set)
    standev = statistics.stdev(Data_set)
    for i in range(12):
        cutoffs[i] = mean + (standev * regions[i])
    for i in range(len(Data_set)):
        if (Data_set[i] < (mean + (standev * regions[0]))):
            distribution[0] = distribution[0] + 1
        elif(Data_set[i] < (mean + (standev * regions[1]))):
            distribution[1] = distribution[1] + 1
        elif(Data_set[i] < (mean + (standev * regions[2]))):
            distribution[2] = distribution[2] + 1
        elif(Data_set[i] < (mean + (standev * regions[3]))):
            distribution[3] = distribution[3] + 1
        elif(Data_set[i] < (mean + (standev * regions[4]))):
            distribution[4] = distribution[4] + 1
        elif(Data_set[i] < (mean + (standev * regions[5]))):
            distribution[5] = distribution[5] + 1
        elif(Data_set[i] < (mean + (standev * regions[6]))):
            distribution[6] = distribution[6] + 1
        elif(Data_set[i] < (mean + (standev * regions[7]))):
            distribution[7] = distribution[7] + 1
        elif(Data_set[i] < (mean + (standev * regions[8]))):
            distribution[8] = distribution[8] + 1
        elif(Data_set[i] < (mean + (standev * regions[9]))):
            distribution[9] = distribution[9] + 1
        elif(Data_set[i] < (mean + (standev * regions[10]))):
            distribution[10] = distribution[10] + 1
        elif(Data_set[i] < (mean + (standev * regions[11]))):
            distribution[11] = distribution[11] + 1
        else:
            distribution[12] = distribution[12] + 1
    for i in range(13):
        distribution[i] = distribution[i] / len(Data_set)
    return distribution, cutoffs
```

```python
# This function checks to make sure a given data set has the proper distribution
def distribution_check(self,Serve_List,variable):
    problem = False 
    length = len(Serve_List)
    Serve_element = [0] * length
    for i in range(length):
        Serve_element[i] = Serve_List[i][variable]
    Train_dist = self.dists[variable]
    Train_cut = self.cutoffs[variable]
    Serv_dist = [0] * 13
    for i in range(length):
        if (Serve_element[i] < Train_cut[0]):
            Serv_dist[0] = Serv_dist[0] + 1
        elif(Serve_element[i] < Train_cut[1]):
            Serv_dist[1] = Serv_dist[1] + 1
        elif(Serve_element[i] < Train_cut[2]):
            Serv_dist[2] = Serv_dist[2] + 1
        elif(Serve_element[i] < Train_cut[3]):
            Serv_dist[3] = Serv_dist[3] + 1
        elif(Serve_element[i] < Train_cut[4]):
            Serv_dist[4] = Serv_dist[4] + 1
        elif(Serve_element[i] < Train_cut[5]):
            Serv_dist[5] = Serv_dist[5] + 1
        elif(Serve_element[i] < Train_cut[6]):
            Serv_dist[6] = Serv_dist[6] + 1
        elif(Serve_element[i] < Train_cut[7]):
            Serv_dist[7] = Serv_dist[7] + 1
        elif(Serve_element[i] < Train_cut[8]):
            Serv_dist[8] = Serv_dist[8] + 1
        elif(Serve_element[i] < Train_cut[9]):
            Serv_dist[9] = Serv_dist[9] + 1
        elif(Serve_element[i] < Train_cut[10]):
            Serv_dist[10] = Serv_dist[10] + 1
        elif(Serve_element[i] < Train_cut[11]):
            Serv_dist[11] = Serv_dist[11] + 1
        else:
            Serv_dist[12] = Serv_dist[12] + 1
    for i in range(13):
        Serv_dist[i] = Serv_dist[i] / length
    a = Train_dist
    b = Serv_dist
    score = (abs(a[0] - b[0])) + (abs(a[1] - b[1])) + (abs(a[2] - b[2])) + (abs(a[3] - b[3])) + (abs(a[4] - b[4])) + (abs(a[5] - b[5])) + (abs(a[6] - b[6])) + (abs(a[7] - b[7])) + (abs(a[8] - b[8])) + (abs(a[9] - b[9])) + (abs(a[10] - b[10])) + (abs(a[11] - b[11])) + (abs(a[12] - b[12]))
    if score > self.dist_lim[variable]:
        print("The Distribution of",self.variable_names[variable],"does not align with a value of", score)
        problem = True
    return problem
```

```python
# This is an intermediate function
def get_roc(Data_set,variable):
    roc_list = []
    columns, rows = Data_set.shape
    for j in range(columns-1):
        temp_val = Data_set[j+1][variable] - Data_set[j][variable]
        roc_list.append(temp_val)
    return roc_list
```

```python
def assign_domain(self,lower,upper,variable):
    self.Domain_lower[variable] = lower
    self.Domain_upper[variable] = upper
```

```python
def assign_agreement(self,lower,upper,agree):
    self.agree_lower = lower
    self.agree_upper = upper
    self.agree_num = int(agree)
```

```python
def assign_distribution_limit(self,limit,variable):
    self.dist_lim[variable] = limit
```

```python
def assign_variable_name(self,name,variable):
    self.variable_names[variable] = name
```

```python
def assign_momentum(self,lower,upper,variable):
    self.momentum_lower[variable] = lower
    self.momentum_upper[variable] = upper
```

```python
def assign_distribution(self,dists,variable):
    self.dists[variable] = dists
```

```python
def assign_distribution_cutoffs(self,cutoffs,variable):
    self.cutoffs[variable] = cutoffs
```

```python
# This function will display all the data entered into including if that
# data is an array. 
def display_data(self,Data_set):
    colors = ['b','g','r','c','m','y']
    columns, rows = Data_set.shape
    fig, axs = plt.subplots(2, self.variables_num)
    axs[0, 0].set(ylabel = 'Original Data')
    axs[1, 0].set(ylabel = 'Comparison Data')
    a = 0
    for i in range(rows):
        a = a + 1
        if (a > 5):
            a = 0
        tempList = []
        for j in range(columns):
            temp = Data_set[j,i]
            tempList.append(temp)
        axs[0, i].plot(self.variable_array[i], colors[a])
        axs[0, i].set_title(self.variable_names[i])
        axs[1, i].plot(tempList, colors[a])
    plt.show()
    return
```

```python
# This function compiles all the previous fucntion into one master function. Each of the
# functions returns boolean value. If the it returns true then that means an alert was raised
# causing the data to be displayed.
def raise_alert(self,Data_set):
    problem = []
    columns, rows = Data_set.shape
    for i in range(rows):
        problem.append(self.domain_check(Serve_List,i))
    for i in range(rows):
        problem.append(self.distribution_check(Serve_List,i))
    if (self.variables_num > 1):
        problem.append(self.agreement_check(Serve_List))
    for i in range(rows):
        problem.append(self.momentum_check(Serve_List,i))
    dis = False
    for i in range(len(problem)):
        if (problem[i] == True):
            dis = True
    if (dis == True):
        self.display_data(Data_set)
        d = 5
    else:
        print("All checks are satisfied")
    return
```

```python
def initialize(self,Data_set):
    columns, rows = Data_set.shape
    self.variables_num = rows
    self.data_length = columns
    for i in range(rows):
        temp = get_roc(Data_set,i)
        self.roc_data.append(temp)
    
    # Organizes data set and names variables
    for i in range(rows):
        tempList = []
        for j in range(columns):
            temp = Data_set[j,i]
            tempList.append(temp)
        self.variable_array.append(tempList)
        tempString = "Variable "
        tempString = tempString + str(i + 1)
        self.variable_names.append(tempString)
        
    # Sets Domain ranges
    for i in range(rows):
        tempDev = statistics.stdev(self.variable_array[i])
        tempMean = statistics.mean(self.variable_array[i])
        self.Domain_lower.append(tempMean - (tempDev*4))
        self.Domain_upper.append(tempMean + (tempDev*4))

    # Sets the agreement ranges
    self.agree_num = rows 
    agreement = 0
    for i in range(columns - 1):
        pos = 0
        neg = 0
        for j in range(rows):
            if (self.roc_data[j][i] > 0):
                pos = pos + 1
            elif (self.roc_data[j][i] < 0):
                neg = neg + 1
        if ((pos >= self.agree_num) or (neg >= self.agree_num)):
            agreement = agreement + 1
    agreement_level = agreement / columns
    self.agree_lower = agreement_level - .1
    self.agree_upper = agreement_level + .1
    
    # Sets momentum ranges
    for j in range(rows):
        momentum = 0
        for i in range(columns - 2):
            if ((self.roc_data[j][i] > 0) and (self.roc_data[j][i+1] > 0)):
                momentum = momentum + 1
            elif ((self.roc_data[j][i] < 0) and (self.roc_data[j][i+1] < 0)):
                momentum = momentum + 1
        mom_level = momentum / (columns - 1)
        self.momentum_lower.append(mom_level - .1)
        self.momentum_upper.append(mom_level + .1)
    
    # Sets the distribution values and cutoff values
    for i in range(rows):
        temp_dist, temp_cut = get_distribution(self.variable_array[i])
        self.dists.append(temp_dist)
        self.cutoffs.append(temp_cut)
        self.dist_lim.append(.8)
```
